```python
# Start with the original salary
df['new_salary'] = df['salary']  # initialize with original salary

# Apply 10% hike for IT
df.loc[df['department'] == 'IT', 'new_salary'] = df['salary'] * 1.10

# Apply 5% hike for HR
df.loc[df['department'] == 'HR', 'new_salary'] = df['salary'] * 1.05
df['new_salary'] = df['new_salary'].round(2)
df
```

```python
df = pd.read_csv('employee.csv')
df1 = pd.read_csv('department.csv')
```

```python
df = df.merge(df1, on='department', how='left')
df.drop('department', axis=1, inplace=True)
df.rename(columns={'department_id': 'department'}, inplace=True)

print(df.head())
df.to_csv('employee_updated.csv', index=False)
```

```python
# Count employees per department
dept_counts = df.groupby('department').size()

print(dept_counts)
```

```python

bins = [0, 20, 30, 40, 50, 60, 100]  
labels = ['0-19', '20-29', '30-39', '40-49', '50-59', '60+']

df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)
age_counts = df['age_group'].value_counts(sort=False)
print(age_counts)
```

```python
plt.figure(figsize=(8,5))
plt.bar(age_counts.index, age_counts.values, color='skyblue')
plt.xlabel('Age Group')
plt.ylabel('Number of People')
plt.title('Number of People in Each Age Group')
plt.show()
```
